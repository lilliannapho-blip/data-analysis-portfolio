```python
import pandas as pd
import sqlite3
license_file = "C:\\users\\thy phan\\Downloads\\license_data.csv"

license_df = pd.read_csv(license_file)
```

```python
try:
    sqlite3_version = sqlite3.sqlite_version
    print(f"SQLite is available, version: {sqlite3_version}")
except ImportError:
    print("SQLite is not installed or not available in this Python environment")

connection = sqlite3.connect("search.db")
cursor = connection.cursor()
license_df.to_sql("License_Data", connection, if_exists="replace", index=False)
```

```python

def search_by_state_county():
    county = input("Enter the county to search: ")
    query = "SELECT month, year, county, prior_state, COUNT(prior_state) AS state_count FROM License_Data WHERE county LIKE ? AND prior_state <> 'Not Applicable' GROUP BY county, prior_state ORDER BY state_count DESC"
    results = cursor.execute(query, ('%' + county + '%',)).fetchall()
    county_df = pd.DataFrame(results, columns=['month', 'year', 'county', 'prior_state', 'state_count'])
    print(county_df)
    
   

def search_by_country_county():
    county2 = input("Enter the county to search: ")
    query = "SELECT month, year, county, prior_country, COUNT(prior_country) AS country_count FROM License_Data WHERE county LIKE ? AND prior_country <> 'USA' GROUP BY county, prior_country ORDER BY country_count DESC"
    results = cursor.execute(query, ('%' + county2 + '%',)).fetchall()
    county2_df = pd.DataFrame(results, columns=['month', 'year', 'county', 'prior_country', 'country_count'])
    print(county2_df)

    

    

def search_by_state():
    state = input("Enter the state to search: ")
    query = "SELECT prior_state, county, COUNT(county) AS county_count FROM License_Data WHERE prior_state LIKE ? GROUP BY county ORDER BY county_count DESC"
    results = cursor.execute(query, ('%' + state + '%',)).fetchall()
    state_df = pd.DataFrame(results, columns=['prior_state', 'county', 'county_count'])
    print(state_df.head(11))

    query2 = "SELECT month, year, prior_state, COUNT(month) AS Month_total, COUNT(year) AS year_total FROM License_Data Where prior_state LIKE? GROUP BY month, year ORDER BY Month_total DESC, year_total DESC"
    results2 = cursor.execute(query2, ('%' + state + '%',)).fetchall()
    state3_df = pd.DataFrame(results2, columns=['month', 'year', 'prior_state', 'Month_total', 'year_total'])
    print(state3_df)
    

def search_by_country_state():
    state2 = input("Enter the state to search: ")
    query = "SELECT prior_state, prior_country, county, COUNT(prior_country) AS country_count, COUNT(county) AS county_count FROM License_Data WHERE prior_state LIKE ? GROUP BY prior_country ORDER BY country_count DESC"
    results = cursor.execute(query, ('%' + state2 + '%',)).fetchall()
    state2_df = pd.DataFrame(results, columns=['prior_state', 'prior_country', 'county', 'country_count', 'county_count'])
    print(state2_df.head(11))

    query2 = "SELECT month, year, prior_state, COUNT(month) AS Month_total, COUNT(year) AS year_total FROM License_Data Where prior_state LIKE? GROUP BY month, year ORDER BY Month_total DESC, year_total DESC"
    results2 = cursor.execute(query2, ('%' + state2 + '%',)).fetchall()
    state3_df = pd.DataFrame(results2, columns=['month', 'year', 'prior_state', 'Month_total', 'year_total'])
    print(state3_df)

    total_country_count = state2_df['country_count'].sum()
    print("Total Out of Country")
    print(total_country_count)  
    
def search_by_country():
    country = input("Enter the country to search: ")
    query = "SELECT prior_country, county, COUNT(county) AS county_count FROM License_Data WHERE prior_country LIKE ? AND prior_country <> 'USA' GROUP BY county ORDER BY county_count DESC"
    results = cursor.execute(query, ('%' + country + '%',)).fetchall()
    country_df = pd.DataFrame(results, columns=['prior_country', 'county', 'county_count'])
    print(country_df)

    query2 = "SELECT month, year, prior_country, COUNT(month) AS Month_total, COUNT(year) AS year_total FROM License_Data Where prior_state LIKE? GROUP BY month, year ORDER BY Month_total DESC, year_total DESC"
    results2 = cursor.execute(query2, ('%' + country + '%',)).fetchall()
    state3_df = pd.DataFrame(results2, columns=['month', 'year', 'prior_country', 'Month_total', 'year_total'])
    print(state3_df)

   

def search_by_state_country():
    country2 = input("Enter the country to search: ")
    query = "SELECT prior_country, prior_state, COUNT(prior_state) AS state_count FROM License_Data WHERE prior_country LIKE ? GROUP BY prior_state, prior_country ORDER BY state_count DESC"
    results = cursor.execute(query, ('%' + country2 + '%',)).fetchall()
    country2_df = pd.DataFrame(results, columns=['prior_country', 'prior_state', 'state_count'])
    print(country2_df)

    query3 = "SELECT month, year, prior_country, COUNT(month) AS Month_total, COUNT(year) AS year_total FROM License_Data Where prior_country LIKE? GROUP BY month, year ORDER BY Month_total DESC, year_total DESC"
    results3 = cursor.execute(query3, ('%' + country2 + '%',)).fetchall()
    state4_df = pd.DataFrame(results3, columns=['month', 'year', 'prior_country', 'Month_total', 'year_total'])
    print(state4_df)

    total_state_count = country2_df['state_count'].sum()
    print("Total Out of State")
    print(total_state_count)

def exit_menu():
    print("Okay! Thank you for visiting today, goodbye.")
    connection.close()
    exit()


def menu():
    print("Hello! What may I call you by?")
    Full_Name = str(input("Enter your full name: "))
    print()
    print("Nice to meet you, {}!".format(Full_Name))
    print()

    
    print("Would you like to search States or Countries?")
    choice = input (" (State/Country): ")
    if choice == "State":
        
        while True:
            print("\n--- Search Engine Menu ---")
            print("1. Search by County")
            print("2. Search by Prior_State")
            print("3. Search by Prior_Country")
            print("4. Exit")

            choice = input("Enter your choice (1, 2, 3, or 4): ")

            if choice == "1":
                search_by_state_county()
            elif choice == "2":
                search_by_state()
            elif choice == "3":
                search_by_state_country()
            elif choice == "4":
                exit_menu()
                break
            else:
                print("Invalid choice. Please try again.")
 
        print("Would you like to search again?")
        choice = input (" (yes/no): ")
        if choice == "yes":
            menu()
        else : 
            print("Okay! Thank you for visiting today, goodbye.")
    
    elif choice == "Country" :
       
        while True:
            print("\n--- Search Engine Menu ---")
            print("1. Search by County")
            print("2. Search by Prior_State")
            print("3. Search by Prior_Country")
            print("4. Exit")

            choice = input("Enter your choice (1, 2, 3, or 4): ")

            if choice == "1":
                search_by_country_county()
            elif choice == "2":
                search_by_country_state()
            elif choice == "3":
                search_by_country()
            elif choice == "4":
                exit_menu()
                break
            else:
                print("Invalid choice. Please try again.")
                
        print("Would you like to search again?")
        choice = input (" (yes/no): ")
        if choice == "yes":
            menu()
        else : 
            print("Okay! Thank you for visiting today, goodbye.")
        
    else :
        print("Okay! Thank you for visiting today, goodbye.")

        
    
menu()

```
